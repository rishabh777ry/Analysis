Customer Segmentation using Python

```python
# Importing necessary Library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_excel('Online Retail.xlsx')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df_clean = df.dropna(subset = ['Description' , 'CustomerID'])
df_clean
```

```python
df_clean.info()
```

```python
total_customer = df_clean['CustomerID'].nunique()
print("Total number of Customer : " , total_customer ) 
```

```python
total_country = df_clean['Country'].nunique()
print('Total Number of Country :', total_country)
```

```python
total_product = df_clean['StockCode'].nunique()
print('Total Number of Products :', total_product)
```

```python
total_bill = df_clean['InvoiceNo'].nunique()
print('Toatal Number of Bills :', total_bill)
```

```python
df_clean.columns
```

```python
plt.figure(figsize=(12, 6))
invoices_per_customer = df_clean.groupby('CustomerID')['InvoiceNo'].nunique()
invoices_per_customer.plot(kind='bar', color='orange')
plt.title('Total Number of Invoices per Customer')
plt.xlabel('CustomerID')
plt.ylabel('Number of Invoices')
```

```python
plt.figure(figsize=(12, 6))
invoices_per_country = df_clean.groupby('Country')['InvoiceNo'].nunique()
invoices_per_country.plot(kind='bar', color='green')
plt.title('Total Number of Invoices per Country')
plt.xlabel('Country')
plt.ylabel('Number of Invoices')
plt.xticks(rotation=45)
```

```python
plt.figure(figsize=(12, 6))
plt.pie(invoices_per_country, labels=invoices_per_country.index, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart of Total Number of Invoices per Country')
plt.axis('equal') 
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Country', y='Quantity', data=df_clean)
plt.title('Box Plot of Quantity by Country')
plt.xlabel('Country')
plt.ylabel('Quantity')
plt.xticks(rotation=45)
```

```python
df_clean.columns
```

```python
customer_per_country = df_clean.groupby('Country')['CustomerID'].nunique()
customer_per_country
```

```python
customer_per_country.plot(kind='bar', color='green')
plt.title('Number of Customers in Each Country')
plt.xlabel('Country')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45)
```

```python
df_clean['Revenue'] = df_clean['Quantity'] * df_clean['UnitPrice'] 
total_sales_per_country = df_clean.groupby('Country')['Revenue'].sum().sort_values(ascending=False)
total_sales_per_country.plot(kind='bar', color='blue')
plt.title('Total Sales per Country')
plt.xlabel('Country')
plt.ylabel('Total Sales (in currency)')
plt.xticks(rotation=45)
```

```python
revenue_by_date = df_clean.groupby('InvoiceDate')[['Revenue']].sum().reset_index()


# Plot the data
revenue_by_date.plot(kind = 'bar')
plt.xlabel('Date')
plt.ylabel('Total Revenue')
plt.title('Total Revenue Over Time')
```
